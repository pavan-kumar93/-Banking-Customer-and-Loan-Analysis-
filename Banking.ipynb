```python
# converting the estimated income( numerical column ) to categorical column as income band
bins = [0,100000,300000,float('inf')]
labels = ['Low','medium','High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins = bins,labels = labels,right = False)
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

```python
# Examine the distribution of unique categories in catgories columns
categorical_cols = df[["BRId","GenderId","Amount of Credit Cards","Nationality","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols :
    print(f"value counts for'{col}':")
    display(df[col].value_counts())
    

```

```python
# univariate analysis of the all the categorical columns
for i, predictor in enumerate(categorical_cols):
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=predictor)
    plt.title(predictor)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

         
```

```python
# # bivariate analysis
for i, predictor in enumerate(categorical_cols):
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=predictor,hue = 'GenderId',palette = 'Set2')
    plt.title(predictor)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Univariate analysis of numerical columns using histogram
numerical_cols = df[["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Saving Accounts","Checking Accounts","Foreign Currency Account","Business Lending"]].columns

plt.figure(figsize=(22,16), dpi=120)

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i+1)   # 3x3 grid instead of 4x3
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col, fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.xticks(rotation=45)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
# Creating a heatmap to find corelation between variables

numerical_cols = df[["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Saving Accounts","Checking Accounts","Foreign Currency Account","Business Lending"]].columns

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize = (6,6))
sns.heatmap(correlation_matrix,annot = True,cmap='crest',fmt=".2f")
plt.title("Correlation matrix")
plt.show()



```
